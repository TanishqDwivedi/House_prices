### Dealing with missing data

We used 2 method to deal with the missing values which are present in the dataset.

<b>1.Drop Missing values:</b>

        If a feature contains more that 50% of the missing value then it better to drop that column as it will have very less or negetive impact on our model accuracy. 

<b>2.Filling Missing values:</b>

        In this project we are taking the most frequent value of the feature and we are filling that value wherever we have msiing value  

For Train Data

```python
dropcol=[] #Dropping the columns having missing values more than 500
for col in train.columns:
    if train[col].isna().sum()>500:
        dropcol.append(col)
print(dropcol)
```

```python
def outlier_detect(train):
    for i in train.describe().columns:
        Q1=train.describe().at['25%',i]
        Q3=train.describe().at['75%',i]
        IQR=Q3 - Q1
        LTV=Q1 - 1.5 * IQR
        UTV=Q3 + 1.5 * IQR
        x=np.array(train[i])
        p=[]
        for j in x:
            if j < LTV or j>UTV:
                p.append(train[i].median())
            else:
                p.append(j)
        train[i]=p
    return train
```

```python
outlier_detect(train)
```

```python
def outlier_detect(test):
    for i in test.describe().columns:
        Q1=test.describe().at['25%',i]
        Q3=test.describe().at['75%',i]
        IQR=Q3 - Q1
        LTV=Q1 - 1.5 * IQR
        UTV=Q3 + 1.5 * IQR
        x=np.array(test[i])
        p=[]
        for j in x:
            if j < LTV or j>UTV:
                p.append(test[i].median())
            else:
                p.append(j)
        test[i]=p
    return test
```

### Feature Selection

**Curse of dimentionality:- model with the huge dimentionliy
with 500-600 feature is less likly to perform well and the 
accuracy will be lower therefore it is very important to reduce and keep only those features who will improve models accuracy. In this project we have applied few feature selection techniques which are as follows**

### 1. Variance

**Dropping constant feature.
If the feature is mostly same value, model is not going to learn anything from it**

```python
from sklearn.feature_selection import VarianceThreshold
```

### 2.Correlation of featuers with each other
**If two of your feature is highly corelated, you can drop one
because they are redundent**

```python
df_corr = train.drop('SalePrice', 1) #Dropping the target column
correlation_mat = df_corr.corr()
corr_pairs = correlation_mat.unstack()
sorted_pairs = corr_pairs.sort_values(kind="quicksort")
strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.6]
strong_pairs = strong_pairs.reset_index()
strong_pairs = strong_pairs.rename(columns={"level_0": "Feature1", "level_1": "Feature2", 0: "Value"})
strong_pairs
```

```python
strong_pairs = strong_pairs.drop_duplicates(subset=['Value'])
strong_pairs = strong_pairs[:-1]
strong_pairs = strong_pairs.reset_index()
strong_pairs
```

```python
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result
```
